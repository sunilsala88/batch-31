```python
import pandas as pd
df3=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
df3['Date added']=pd.to_datetime(df3['Date added'])
df3.info()
```

```python
def fun1(d1):
    return d1.year+d1.month+d1.day
df3['new']=df3['Date added'].apply(fun1)
df3
```

```python
import math
def fun1(d1):
    return math.sqrt(d1)
df3['new']=df3['CIK'].apply(fun1)
df3
```
